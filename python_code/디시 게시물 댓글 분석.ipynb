```python
datetime_format = "%Y.%m.%d %H:%M:%S"
res = []
for i in range(0, 3164):
  x = df['날짜'][i]
  dre = datetime.strptime(x, datetime_format)
  res.append(dre)

df['date'] = res
```

```python
df['year'] = df['날짜'].dt.year
```

```python
df.to_csv("/content/지하철시체1.csv", index=False, encoding='utf-8-sig')
```
